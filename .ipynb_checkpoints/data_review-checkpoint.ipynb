**b) Extreme Values and Outliers** <br>
The points shows drastic changes over the years, resulting in heavy distribution. This is majorly due to changes in voting system, shown in [Voting at the Eurovision Song Contest](https://en.wikipedia.org/wiki/Voting_at_the_Eurovision_Song_Contest).
Spike in year 1971 is caused by changed voting system to distirbutes 10 points from 10 to 1 to favorites songs, from distribute 9 points juries got as a total. Doubled points in year 2016 is due to adopting telepoints.
The fall in year 1974 is also from reverting voting system to distribute 10 points juries got as a total, from changes in 1971.<br>
The places over the years showed changes among the years. Overall the places are getting stable without ties as contest goes on, having only 1 song per 1 place. In the year 1956, no points nor place was publicated, all placed in 2nd. In year 1969, 4 countries are tied and got 1st place causing changes in voting system to prevent ties for winners. <br>
To address these distributions, we need measures to normalize points and deal with tied places.

# Clustering

```python
#Load Data
contestant_df = pd.read_csv('dataset/contestants_cleaned.csv')
vote_df = pd.read_csv('dataset/votes_cleaned.csv')
```

```python
# Filter the data for `round = final`
final_round_votes = vote_df[vote_df['round'] == 'final']

# Pivot table to create the voting matrix using `perc_of_max`
voting_matrix_perc_of_max = final_round_votes.pivot_table(
    index='from_country_id',
    columns='to_country_id',
    values='perc_of_max',
    aggfunc='sum',
    fill_value=0
)

# To display the voting matrix nicely in Jupyter Notebook
import pandas as pd
# In Jupyter Notebook: use `.style.format()` for better formatting
voting_matrix_perc_of_max.style.format(precision=2)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Reduce dimensions to 2D using PCA
pca = PCA(n_components=2)
voting_matrix_2d = pca.fit_transform(voting_matrix_perc_of_max)

# Apply K-Means clustering (choosing an arbitrary number of clusters, e.g., 5)
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(voting_matrix_perc_of_max)

# Create a scatterplot of the clusters
plt.figure(figsize=(10, 8))
scatter = plt.scatter(voting_matrix_2d[:, 0], voting_matrix_2d[:, 1], c=clusters, cmap='viridis', s=100, alpha=0.8)
plt.colorbar(scatter, label='Cluster ID')
plt.title('Country Clustering Based on Voting Patterns')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Annotate each point with the country code
for i, country in enumerate(voting_matrix_perc_of_max.index):
    plt.annotate(country, (voting_matrix_2d[i, 0], voting_matrix_2d[i, 1]), fontsize=8, alpha=0.7)

plt.show()
```

```python
# Invert the voting matrix to emphasize non-voting patterns
# Any value above a threshold (e.g., 0) will be set to 0, and non-voting relationships will be emphasized
non_voting_matrix = voting_matrix_perc_of_max.applymap(lambda x: 0 if x > 0 else 1)

# Reduce dimensions to 2D using PCA
non_voting_matrix_2d = PCA(n_components=2).fit_transform(non_voting_matrix)

# Apply K-Means clustering to the non-voting matrix
kmeans_non_voting = KMeans(n_clusters=5, random_state=42)
clusters_non_voting = kmeans_non_voting.fit_predict(non_voting_matrix)

# Create a scatterplot for non-voting clustering
plt.figure(figsize=(10, 8))
scatter_non_voting = plt.scatter(non_voting_matrix_2d[:, 0], non_voting_matrix_2d[:, 1], c=clusters_non_voting, cmap='coolwarm', s=100, alpha=0.8)
plt.colorbar(scatter_non_voting, label='Cluster ID (Non-Voting)')
plt.title('Country Clustering Based on Non-Voting Patterns')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Annotate each point with the country code
for i, country in enumerate(non_voting_matrix.index):
    plt.annotate(country, (non_voting_matrix_2d[i, 0], non_voting_matrix_2d[i, 1]), fontsize=8, alpha=0.7)

plt.show()
```
